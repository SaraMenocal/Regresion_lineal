```python
# Librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cdmx_sin_nas.csv")
# Eliminamos la columna Unnamed: 0 que se creó al guardar el csv
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.shape
```

```python
# Checamos NAs
print(df.isnull().sum().to_frame(name='nulos').to_string())
```

```python
#Se separa en 2 dataframe para poder procesar los outliers
# Separar por tipo de dato
cuantitativas = df.select_dtypes(include=["number"])
cualitativas  = df.select_dtypes(exclude=["number"])
```

```python
# Se decidio utilizar el método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuantitativas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
data_iqr= cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_iqr.info()
```

```python
data_iqr.isnull().sum()
```

```python
# Sutituir NAs de columnas tipo float con la media 
data_iqr['latitude'] = data_iqr['latitude'].fillna(round(data_iqr['latitude'].mean()))
data_iqr['longitude'] = data_iqr['longitude'].fillna(round(data_iqr['longitude'].mean()))
data_iqr['bathrooms'] = data_iqr['bathrooms'].fillna(round(data_iqr['bathrooms'].mean(),1))
data_iqr['review_scores_rating'] = data_iqr['review_scores_rating'].fillna(round(data_iqr['review_scores_rating'].mean()))
data_iqr['review_scores_accuracy'] = data_iqr['review_scores_accuracy'].fillna(round(data_iqr['review_scores_accuracy'].mean()))
data_iqr['review_scores_cleanliness'] = data_iqr['review_scores_cleanliness'].fillna(round(data_iqr['review_scores_cleanliness'].mean()))
data_iqr['review_scores_checkin'] = data_iqr['review_scores_checkin'].fillna(round(data_iqr['review_scores_checkin'].mean()))
data_iqr['review_scores_communication'] = data_iqr['review_scores_communication'].fillna(round(data_iqr['review_scores_communication'].mean()))
data_iqr['review_scores_location'] = data_iqr['review_scores_location'].fillna(round(data_iqr['review_scores_location'].mean()))
data_iqr['review_scores_value'] = data_iqr['review_scores_value'].fillna(round(data_iqr['review_scores_value'].mean()))
data_iqr['reviews_per_month'] = data_iqr['reviews_per_month'].fillna(round(data_iqr['reviews_per_month'].mean()))
data_iqr['minimum_nights_avg_ntm'] = data_iqr['minimum_nights_avg_ntm'].fillna(round(data_iqr['minimum_nights_avg_ntm'].mean(),1))
data_iqr['maximum_nights_avg_ntm'] = data_iqr['maximum_nights_avg_ntm'].fillna(round(data_iqr['maximum_nights_avg_ntm'].mean(),1))
#data_iqr[] = data_iqr[].fillna(round(data_iqr[].mean()))
```

```python
# Sustituir NAs de columnas tipo int con la mediana
data_iqr['host_response_rate'] = data_iqr['host_response_rate'].fillna(round(data_iqr['host_response_rate'].median()))
data_iqr['host_acceptance_rate'] = data_iqr['host_acceptance_rate'].fillna(round(data_iqr['host_acceptance_rate'].median()))
data_iqr['host_listings_count'] = data_iqr['host_listings_count'].fillna(round(data_iqr['host_listings_count'].median()))
data_iqr['host_total_listings_count'] = data_iqr['host_total_listings_count'].fillna(round(data_iqr['host_total_listings_count'].median()))
data_iqr['accommodates'] = data_iqr['accommodates'].fillna(round(data_iqr['accommodates'].median()))
data_iqr['bedrooms'] = data_iqr['bedrooms'].fillna(round(data_iqr['bedrooms'].median()))
data_iqr['beds'] = data_iqr['beds'].fillna(round(data_iqr['beds'].median()))
data_iqr['price'] = data_iqr['price'].fillna(round(data_iqr['price'].median()))
data_iqr['minimum_nights'] = data_iqr['minimum_nights'].fillna(round(data_iqr['minimum_nights'].median()))
data_iqr['minimum_minimum_nights'] = data_iqr['minimum_minimum_nights'].fillna(round(data_iqr['minimum_minimum_nights'].median()))
data_iqr['maximum_minimum_nights'] = data_iqr['maximum_minimum_nights'].fillna(round(data_iqr['maximum_minimum_nights'].median()))
data_iqr['minimum_maximum_nights'] = data_iqr['minimum_maximum_nights'].fillna(round(data_iqr['minimum_maximum_nights'].median()))
data_iqr['maximum_maximum_nights'] = data_iqr['maximum_maximum_nights'].fillna(round(data_iqr['maximum_maximum_nights'].median()))
data_iqr['number_of_reviews'] = data_iqr['number_of_reviews'].fillna(round(data_iqr['number_of_reviews'].median()))
data_iqr['number_of_reviews_ltm'] = data_iqr['number_of_reviews_ltm'].fillna(round(data_iqr['number_of_reviews_ltm'].median()))
data_iqr['number_of_reviews_l30d'] = data_iqr['number_of_reviews_l30d'].fillna(round(data_iqr['number_of_reviews_l30d'].median()))
data_iqr['number_of_reviews_ly'] = data_iqr['number_of_reviews_ly'].fillna(round(data_iqr['number_of_reviews_ly'].median()))
data_iqr['estimated_revenue_l365d'] = data_iqr['estimated_revenue_l365d'].fillna(round(data_iqr['estimated_revenue_l365d'].median()))
data_iqr['calculated_host_listings_count'] = data_iqr['calculated_host_listings_count'].fillna(round(data_iqr['calculated_host_listings_count'].median()))
data_iqr['calculated_host_listings_count_entire_homes'] = data_iqr['calculated_host_listings_count_entire_homes'].fillna(round(data_iqr['calculated_host_listings_count_entire_homes'].median()))
data_iqr['calculated_host_listings_count_private_rooms'] = data_iqr['calculated_host_listings_count_private_rooms'].fillna(round(data_iqr['calculated_host_listings_count_private_rooms'].median()))
data_iqr['calculated_host_listings_count_shared_rooms'] = data_iqr['calculated_host_listings_count_shared_rooms'].fillna(round(data_iqr['calculated_host_listings_count_shared_rooms'].median()))
```
