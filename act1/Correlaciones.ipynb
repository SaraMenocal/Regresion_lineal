```python
# Librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Función de Top 10 correlaciones 
def top_correlations(df, tipo_habitacion, top_n=10):
    corr_matrix = df.corr(numeric_only=True)   #.... matriz de correlacion solo con variables numericas 
    corr_sums = corr_matrix.abs().sum().sort_values(ascending=False)  # .... suma de correlaciones absolutas por variables
    top_vars = corr_sums.head(top_n).index   # .... seleccion de top 10 correlaciones 
    top_corrs = corr_matrix.loc[top_vars, top_vars] # Extracción de la submatriz de correlaciones
    print(f"Top 10 variables más correlacionadas para: {tipo_habitacion}")
    print(top_corrs.round(4))

    # Heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(top_corrs, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f"Heatmap de correlaciones - {tipo_habitacion}")
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Se aplica la función a cada tipo de alojamiento
top_correlations(entire_home, "Entire Home")
top_correlations(private_room, "Private Room")
top_correlations(shared_room, "Shared Room")
top_correlations(hotel_room, "Hotel Room")
```
