```python
# Librerías utilizadas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Mexico_City.csv') # Cargar ya el dataset limpio de NAs y Outliers
df.shape
```

```python
# Dividir el dataset por tipo de cuarto (4 subsets)
entire_home = df[df['room_type'] == 'Entire home/apt'].copy()
private_room = df[df['room_type'] == 'Private room'].copy()
shared_room = df[df['room_type'] == 'Shared room'].copy()
hotel_room = df[df['room_type'] == 'Hotel room'].copy()
```

```python
# Análisis de correlación para cada subconjunto de 6 pares de variables
# 1. host_acceptance_rate vs host_response_rate
# 2. review_scores_rating vs host_listings_count
# 3. host_acceptance_rate vs price
# 4. availability_365 vs number_of_reviews
# 5. host_acceptance_rate vs number_of_reviews
# 6. reviews_per_month vs review_scores_communication
```

**Entire home/apt**


```python
entire_home.shape # (17,235/ 76, un 65% de los datos)
```

```python
corr1 = entire_home['host_acceptance_rate'].corr(entire_home['host_response_rate'])
corr2 = entire_home['review_scores_rating'].corr(entire_home['calculated_host_listings_count'])
corr3 = entire_home['host_acceptance_rate'].corr(entire_home['price'])
corr4 = entire_home['availability_365'].corr(entire_home['number_of_reviews'])
corr5 = entire_home['host_acceptance_rate'].corr(entire_home['number_of_reviews'])
corr6 = entire_home['reviews_per_month'].corr(entire_home['review_scores_communication'])

# Imprimir correlaciones redondeadas
print("Correlaciones:")
print(f"1. host_acceptance_rate vs host_response_rate: {round(corr1, 3)}")
print(f"2. review_scores_rating vs calculated_host_listings_count: {round(corr2, 3)}")
print(f"3. Host_acceptance_rate vs price: {round(corr3, 3)}")
print(f"4. availability_365 vs number_of_reviews: {round(corr4, 3)}")
print(f"5. Host_acceptance_rate vs number_of_reviews: {round(corr5, 3)}")
print(f"6. reviews_per_month vs review_scores_communication: {round(corr6, 3)}")
```

```python
# 1.1. host_acceptance_rate vs host_response_rate 'entire home/apt' (y, x)
corr= entire_home['host_acceptance_rate'].corr(entire_home['host_response_rate'])
print(f"Correlación: {round(corr, 3)}")

# Gráfico de dispersión con línea de regresión lineal ajustada para mostrar la tendencia.
sns.scatterplot(data=entire_home, x='host_response_rate', y='host_acceptance_rate', alpha=0.5)
sns.regplot(data=entire_home, x='host_response_rate', y='host_acceptance_rate', scatter=False, color='orange')
plt.title('Acceptance Rate vs Response Rate (Entire home/apt)')
plt.tight_layout()
plt.show()
```
